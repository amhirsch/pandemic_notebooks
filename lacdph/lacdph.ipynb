```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('LA_County_Covid19_cases_deaths_date_table.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df['date_use'] = pd.to_datetime(df['date_use'])
```

```python
fig, ax = plt.subplots(dpi=150)
sns.lineplot(x='date_use', y='avg_cases', data=df, ax=ax)
ax.set_xlim(pd.to_datetime('2020-12-01'), pd.to_datetime('2021-02-01'))
ax.set_ylim(4_000)
date_form = DateFormatter("%m-%d")
ax.xaxis.set_major_formatter(date_form)
fig.show()
```
